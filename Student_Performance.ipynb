**Q2. What is the general test score distribution of each subject?**

**Inputing the pass marks.**

```python
pass_mark = 45
```

```python
df['math_pass'] = np.where(df['math score']<pass_mark, 'Fail', 'Pass')
df.math_pass.value_counts()
```

```python
df['reading_pass'] = np.where(df['reading score']<pass_mark, 'Fail', 'Pass')
df.reading_pass.value_counts()
```

```python
df['writing_pass'] = np.where(df['writing score']<pass_mark, 'Fail', 'Pass')
df.writing_pass.value_counts()
```

Reading had the highest number of passes and the least failures, while maths had the highest number of failures and the least passes. This means that students performed better on Reading than on Math.

**Q3. What is the relationship between the passrates and the gender?**

```python
plt.figure(figsize=(25,8))

plt.subplot(1, 4, 1)
gender = sns.countplot(x = "gender", hue="math_pass", color='lime', data=df)
plt.title("Rate of Maths Pass between the different Genders")
plt.xlabel("Gender")
plt.ylabel("Math_pass")
_ = plt.setp(gender.get_xticklabels(), rotation=45)

plt.subplot(1, 4, 2)
gender = sns.countplot(x = "gender", hue="writing_pass", color='blue', data=df)
plt.title("Rate of writing Pass between the different Genders")
plt.xlabel("Gender")
plt.ylabel("Writing_pass")
_ = plt.setp(gender.get_xticklabels(), rotation=45)

plt.subplot(1, 4, 3)
gender = sns.countplot(x = "gender", hue="reading_pass", color='orange', data=df)
plt.title("Rate of Reading Pass between the different Genders")
plt.xlabel("Gender")
plt.ylabel("Reading_pass")
_ = plt.setp(gender.get_xticklabels(), rotation=45)
plt.show()
```

From the visualization, there are similar pass rates between both genders in math. On the other hand females did better than males in reading and writing.

More females failed maths than males but more males failed reading and writing than the females.

**Final grading of students by percentage.**

```python
df['Percentage'] = df['Total score']/3

def GetGrade(Percentage):
    if ( Percentage >= 80 ):
        return 'A'
    if ( Percentage >= 70):
        return 'B'
    if ( Percentage >= 60):
        return 'C'
    if ( Percentage >= 50):
        return 'D'
    if ( Percentage >= 40):
        return 'E'
    else:
        return 'F'

df['Grade'] = df.apply(lambda x : GetGrade(x['Percentage']), axis=1)

df.Grade.value_counts()
```

More students scored B and C grades (pass) than those who scored E and F grades (fail).

**Q9. What is the general grade distribution of the students?**
